```python
import cv2  # This is OpenCV, used for image processing like loading and editing images
import supervision as sv  # This is Supervision, a library to help with object detection tasks
from ultralytics import YOLO  # This is YOLO, a model for detecting objects in images
import numpy as np  # NumPy helps with math and handling arrays (like lists of numbers)
```

Load image using opencv


```python
import cv2
from google.colab.patches import cv2_imshow

# Load image
image_path = '/content/IMG_1470_jpg.rf.709477d881ea8a65a1c21dd1894ac967.jpg'  # Replace with the actual path to your image file
original_image = cv2.imread(image_path)

# Check if image was loaded successfully
if original_image is not None:
  # Display the image
  cv2_imshow(original_image)
else:
  print(f"Error: Unable to load image from {image_path}")
```

Resize image to fit into the model

```python
h, w = original_image.shape[:2]  # Get the height (h) and width (w) of the image
new_w = 640  # Set a new width for the resized image
new_h = int(h * (new_w / w))  # Calculate the new height to keep the image's proportions
img = cv2.resize(original_image, (new_w, new_h))  # Resize the image to the new dimensions
```

load model and run inference

```python
# Load the YOLO model and use it to detect objects in the image
model = YOLO("chilliv11overlapcheryl.pt")  # Load the trained YOLO model (best.pt is the model file)
results = model(img, retina_masks=True)[0]  # Run the model on the image to detect objects; retina_masks helps with accurate masks
detections = sv.Detections.from_ultralytics(results)  # Convert YOLO results into a format Supervision can use
```

Get masks and class IDs from the detections


```python
# Get masks and class IDs from the detections
masks = detections.mask  # Masks are like outlines of detected objects (e.g., chilies or disease spots)
class_ids = detections.class_id  # Class IDs tell us what each detected object is (0 for disease, 1 for chili)
```

Find which detections are chilies and which are disease spots


```python
chili_indices = [i for i, cid in enumerate(class_ids) if cid == 1]  # List of indices where the object is a chili
disease_indices = [i for i, cid in enumerate(class_ids) if cid == 0]  # List of indices where the object is a disease spot
```

Set up tools to draw on the image

```python
mask_annotator = sv.PolygonAnnotator()  # Tool to draw outlines (masks) around detected objects
label_annotator = sv.LabelAnnotator()  # Tool to add text labels to detected objects
```

Draw masks and labels on the image

```python
annotated_image = mask_annotator.annotate(scene=img.copy(), detections=detections)  # Draw masks on a copy of the image
annotated_image = label_annotator.annotate(scene=annotated_image, detections=detections)  # Add labels to the image with masks
```

Calculate the percentage of disease on each chili


```python
percentage = 0.0
for idx, chili_idx in enumerate(chili_indices):  # Loop through each detected chili
    chili_mask = masks[chili_idx]  # Get the mask (outline) of the current chili
    chili_area = np.sum(chili_mask)  # Calculate the total area of the chili (count of pixels in the mask)
    disease_area = 0  # Start with zero disease area

    # Check if any disease spots overlap with this chili
    if disease_indices:  # If there are disease spots detected
        combined_disease_mask = np.any([masks[d_idx] for d_idx in disease_indices], axis=0)  # Combine all disease masks into one
        overlap = np.logical_and(chili_mask, combined_disease_mask)  # Find where the chili and disease masks overlap
        disease_area = np.sum(overlap)  # Calculate the area of the overlap (disease on the chili)

    # Calculate the percentage of the chili that is diseased
    if chili_area > 0:  # If the chili has a valid area
        percentage = (disease_area / chili_area) * 100  # Calculate disease percentage
        text = f"{percentage:.2f}%"  # Format the percentage as text (e.g., "12.34%")
    else:
        text = "0.00% (No chili area)"  # If no chili area, show 0%

    # Find the center of the chili to place the text
    moments = cv2.moments(chili_mask.astype(np.uint8))  # Calculate the "center of mass" of the chili mask
    if moments["m00"] != 0:  # If the chili has a valid area
        cX = int(moments["m10"] / moments["m00"])  # X-coordinate of the center
        cY = int(moments["m01"] / moments["m00"])  # Y-coordinate of the center
        cv2.putText(
            annotated_image, text, (cX, cY),  # Add the percentage text at the center
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2, cv2.LINE_AA  # Set font, size, green color, thickness, and style
        )
```

```python
# Show the image with masks and percentage labels
cv2_imshow(annotated_image)  # Display the image in a window
print(f"{percentage:.2f}%")
if percentage <= 2:
  print("score 0")
elif percentage > 2 and percentage <=5:
  print("score 1")
elif percentage > 5 and percentage <=15:
  print("score 3")
elif percentage > 15 and percentage <=25:
  print("score 7")
elif percentage > 25:
  print("score 9")
```

Define a function to classify chilies as infected or healthy


```python
def process_chili_detections(detections):  # Function takes detection results as input
    masks = detections.mask  # Get the masks of detected objects
    class_ids = detections.class_id  # Get the class IDs (0=disease, 1=chili)
    boxes = detections.xyxy  # Get the bounding boxes (rectangles around objects)

    # Find indices for chilies and disease spots
    chili_indices = [i for i, cid in enumerate(class_ids) if cid == 1]  # List of chili indices
    disease_indices = [i for i, cid in enumerate(class_ids) if cid == 0]  # List of disease indices

    # Create lists to store new detection information
    new_boxes = []  # List for new bounding boxes
    new_class_ids = []  # List for new class IDs (2=infected, 3=healthy)
    new_confidence = []  # List for confidence scores

    # Check each chili for disease
    for chili_idx in chili_indices:  # Loop through each chili
        chili_mask = masks[chili_idx]  # Get the chili's mask
        is_infected = False  # Assume the chili is healthy

        # Check if any disease mask overlaps with this chili
        for d_idx in disease_indices:  # Loop through disease spots
            disease_mask = masks[d_idx]  # Get the disease mask
            if np.any(np.logical_and(chili_mask, disease_mask)):  # Check for overlap
                is_infected = True  # Mark chili as infected
                break  # No need to check more disease spots

        # Add the chili's bounding box and new class ID
        new_boxes.append(boxes[chili_idx])  # Keep the chili's bounding box
        new_class_ids.append(2 if is_infected else 3)  # Set class ID to 2 (infected) or 3 (healthy)
        new_confidence.append(detections.confidence[chili_idx])  # Keep the confidence score

    # Create a new detection object with the updated information
    return sv.Detections(
        xyxy=np.array(new_boxes) if new_boxes else np.empty((0, 4)),  # Convert boxes to array or empty array
        class_id=np.array(new_class_ids) if new_class_ids else np.array([]),  # Convert class IDs to array or empty
        confidence=np.array(new_confidence) if new_confidence else np.array([])  # Convert confidence scores to array or empty
    )
```

Process the detections to classify chilies

```python
new_detections = process_chili_detections(detections)  # Get new detections with infected/healthy labels

# Set up tools to draw bounding boxes and labels
box_annotator_bbox = sv.BoxAnnotator()  # Tool to draw rectangles around objects
label_annotator_bbox = sv.LabelAnnotator()  # Tool to add text labels
labels = ["chili", "disease", "infected chili", "healthy chili"]  # List of possible labels

# Draw bounding boxes and labels on a new copy of the image
annotated_image_bbox = box_annotator_bbox.annotate(
    scene=img.copy(),  # Start with a fresh copy of the image
    detections=new_detections  # Use the new detections
)

annotated_image_bbox = label_annotator_bbox.annotate(
    scene=annotated_image_bbox,  # Add labels to the image with boxes
    detections=new_detections,  # Use the new detections
    labels=[labels[class_id] for class_id in new_detections.class_id] if new_detections.class_id.size > 0 else []  # Assign labels based on class IDs
)

# Show the image with bounding boxes and labels
cv2_imshow(annotated_image_bbox)  # Display the image in a window
```
